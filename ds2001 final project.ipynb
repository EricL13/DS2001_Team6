```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
vax = pd.read_csv('/Users/stephaniefissel/Downloads/US_vaccination_data.csv')
vax
```

```python
vax['daily_vaccinations'].isna().sum()
```

```python
vax_drop = vax.drop(["country", "iso_code", "total_vaccinations", "people_vaccinated", "people_fully_vaccinated", "daily_vaccinations_raw", "total_vaccinations_per_hundred", "people_vaccinated_per_hundred", "people_fully_vaccinated_per_hundred", "daily_vaccinations_per_million", "vaccines"], axis = 1)
```

```python
vax1 = vax_drop.rename(columns={"date": "Date"})
```

```python
vax1.Date = pd.to_datetime(vax1.Date)
vax1.set_index('Date', inplace=True)
```

```python
vaxxed = vax1.resample('1M').sum()
```

```python
vaxxed
```

```python
unemployment = pd.read_csv('/Users/stephaniefissel/Downloads/unemployment.csv')
unemployment.head()
```

```python
unemp = unemployment.rename(columns={"Yearmon": "Date", "US UnemploymentRate": "Unemployment Rate" })
```

```python
unemp['Date'] = pd.date_range('01-01-2020', periods=1499, freq='D')
mask = (unemp['Date'] > '01-01-2020') & (unemp['Date'] <= '9-30-2021')
un = (unemp.loc[mask])
un.head()
```

```python
un.set_index('Date', inplace=True)
un.index = pd.to_datetime(un.index)
covid_unemp = un.resample('1M').mean()
```

```python
covid_unemp
```

```python
table1 = pd.concat([vaxxed, covid_unemp], axis=1, join="inner")
table1
```
